```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Assignment4_SalaryData.csv')
print(df)
```

```python
x =df.iloc[:,0].values
y =df.iloc[:,-1].values
print(x)
print(y)
```

```python
ax=df.plot.line(title='Salary V/s Experience Comparesion')
ax.set(xlabel='YearsExperience', ylabel='Salary')
```

```python
index = np.arange(len(x))
plt.bar(index, y)
plt.xlabel('yearsExperience', fontsize=14)
plt.ylabel('salary', fontsize=14)
plt.title('salary vs experience')
plt.show()
```
